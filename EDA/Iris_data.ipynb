Relationship between `sepal-length` and `sepal-width`

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data)
```

Doing a pair-plot: plotting every feature with every other feature

```python
sns.pairplot(data=data, hue='species')
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
data.columns
```

```python
for el in data.columns:
    if(el == 'species'):
        continue
    plot = sns.FacetGrid(data, hue='species')
    plot.map(sns.distplot, el)
    plt.show()
```

```python
p = sns.JointGrid(data=data, x='sepal_length', y='petal_length')
p.plot(sns.regplot, sns.distplot)
```

```python
# 3D scatter plot

import plotly.express as px

fig = px.scatter_3d(data, x='sepal_length', y='petal_length', z='petal_width', color='species')
fig.show()
```
