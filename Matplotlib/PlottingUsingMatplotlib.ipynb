```python
months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
sales1 = [1485, 1218, 1677, 1609, 1534, 1615, 1741, 1689, 1756, 1305, 1452, 1248]
sales2 = [1486, 1068, 1344, 1315, 1811, 1364, 1708, 1900, 1478, 1061, 1912, 1407]
x_1 = np.arange(0, len(months)) - 0.15
x_2 = np.arange(0, len(months)) + 0.15

plt.figure(figsize=(12, 4))
plt.bar(x_1, sales1, width = 0.3, color='red', label='Sales1')
plt.bar(x_2, sales2, width = 0.3, color='blue', label='Sales2')

plt.xticks(np.arange(0, len(months)), labels= months, rotation = 45)
plt.xlabel('Time')
plt.ylabel('Sales Values')
plt.title('Share price over time', size = 18, color = '#0e63eb')
```

### SUBPLOT

```python
x = np.arange(1, 20)
north = np.random.randint(1, 1000, size = len(x))
south = np.random.randint(1, 1000, size = len(x))
east = np.random.randint(1, 1000, size = len(x))
west = np.random.randint(1, 1000, size = len(x))

plt.figure(figsize=(7, 7))

plt.subplot(2,2,1)
plt.plot(x, north, label='north')
plt.legend()

plt.subplot(2,2,2)
plt.plot(x, south, label='south')
plt.legend()

plt.subplot(2,2,3)
plt.plot(x, east, label='east')
plt.legend()

plt.subplot(2,2,4)
plt.plot(x, west, label='west')
plt.legend()

plt.show()
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(10, 10))

axes[0, 0].plot(x, north, label='north')
axes[0, 0].legend()

axes[0, 1].plot(x, south, label='south')
axes[0, 1].legend()

axes[1, 0].plot(x, west, label='west')
axes[1, 0].legend()

axes[1, 1].plot(x, east, label='east')
axes[1, 1].legend()

plt.tight_layout()
```
