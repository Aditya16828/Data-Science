```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./spam.csv', encoding="ISO-8859-1")
data.head()
```

```python
data = data[['v1', 'v2']]
```

```python
data.head()
```

# Removing punctuations

```python
from string import punctuation

def remove_punctuations(str):
    temp = "".join([i for i in str if i not in punctuation])
    return temp
```

```python
data['clean'] = data['v2'].apply(lambda x:remove_punctuations(x))
data.head()
```

# Lowercasing

```python
data['clean_lower'] = data['clean'].apply(lambda x: x.lower())
```

```python
data.head()
```

# Tokenization

```python
import re
```

```python
def tokenize(str):
    tokenized_str = re.findall("[\w]+", str)
    return tokenized_str
```

```python
data['clean_lower_tokenized'] = data['clean_lower'].apply(lambda x:tokenize(x))
data.head()
```

# Remove stopwords

```python
from nltk.corpus import stopwords
```

```python
def remove_stopwords(str_list):
    temp = [ele for ele in str_list if ele not in stopwords.words("english")]
    return temp
```

```python
data['removed_stopwords'] = data['clean_lower_tokenized'].apply(lambda x:remove_stopwords(x))
data.head()
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(str):
    stem_str = [ps.stem(ele) for ele in str]
    return stem_str
```

```python
data['stemmed'] = data['removed_stopwords'].apply(lambda x:stem(x))
data.head()
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()
```

```python
def lemmatize(str_list):
    str_lemmatized = [wnl.lemmatize(ele) for ele in str_list]
    return str_lemmatized
```

```python
data['lemmatized'] = data['removed_stopwords'].apply(lambda x:lemmatize(x))
data.head()
```
